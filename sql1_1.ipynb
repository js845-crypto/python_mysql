```python
import pymysql

conn = pymysql.connect(
    host="localhost",
    user="root",
    password="class1234",
    database="test",
    charset="utf8"
)

cursor = conn.cursor() # 연결된 DB에 SQL을 실행하는 객체

print("MySQL 연결 성공")
```

```python
products = [
    {"name": "키보드", "price": 50000},
    {"name": "마우스", "price": 25000},
    {"name": "모니터", "price": 300000},
    {"name": "노트북", "price": 1200000},
    {"name": "USB", "price": 15000},
    {"name": "웹캠", "price": 80000},
    {"name": "헤드셋", "price": 90000},
    {"name": "스피커", "price": 70000},
    {"name": "외장하드", "price": 140000},
    {"name": "프린터", "price": 220000}
]
```

```python
sql = """
insert into product(name, price)
values(%s, %s)
"""
```

```python
for i, p in enumerate(products):
    # execute(str, tuple) 채울 표현식%가 없으면 두번째 인수는 생략, 하나라도 있으면 튜플로 순서데로 입력
    # ex) (1,) , (1, 1), (1, 1, 1), ...
    cursor.execute(sql, (p["name"], p["price"]))
    print(f"{i+1}번째 insert 실행")

conn.commit()
```

```python
sql = """
select * from product
"""

cursor.execute(sql)
rows = cursor.fetchall()
print(type(rows)) # ( (), (), (), ...)
print(rows)
# mysql에서 삭제하고 나면 이걸 다시 실행하고 아래의 for 문을 실행해야 반영된다.
```

```python
for r in rows:
    print(r)

```

```python
# id가 10인 마우스의 price null -> 150000으로 수정
update_price_val = 150000
update_id = 11
sql = """
update product
set price = %s
where id = %s
"""
cursor.execute(sql, (update_price_val,update_id ))

conn.commit()
print("수정완료")
```

```python
try:

    sql = """
    delete from product
    where price is null
    """ 
# db null값을 연산할 때는 db에서 제공하는 null 연산자 만을 사용해야한다.
# is null, is not null,  .../ = , != tkdyddksehla

    cursor.execute(sql)
    conn.commit()
# select 실행 후 cursor에 반환된 결과셋을 가진다
# insert/ update/ delete는 cursor에 반환된 행의 갯수를 가진다 
    print(f"{cursor.rowcount}행 삭제 성공")
except:
    print("db 예외 발생")

finally:
    cursor.close()
    conn.close()
```
